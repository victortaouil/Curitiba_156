```python
#Primeiro passo importar os dados coletados no site da prefeitura de Curitiba

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np 


df_jan = pd.read_csv('2022-02-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';',encoding = 'latin-1', index_col=False)
df_fev = pd.read_csv('2022-03-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_mar = pd.read_csv('2022-04-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "utf8", index_col=False)
df_abr = pd.read_csv('2022-05-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_mai = pd.read_csv('2022-06-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_jun = pd.read_csv('2022-07-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_jul = pd.read_csv('2022-08-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_ago = pd.read_csv('2022-09-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_setembro = pd.read_csv('2022-10-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_out = pd.read_csv('2022-11-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_nov = pd.read_csv('2022-12-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)
df_dez = pd.read_csv('2023-01-01_156_-_Base_de_Dados.csv', error_bad_lines = False, skiprows = [1], sep =';', encoding = "latin-1", index_col=False)

df_fev.head(6)

#Agora concatenemos todos os DFs em torno da data

df_total = pd.concat([df_jan,df_fev,df_mar,df_abr,df_mai,df_jun,df_jul,df_ago,df_setembro, df_out, df_nov, df_dez], join = 'inner', ignore_index= True)


```

```python
# Vamos identificar as colunas

df_total.columns
df_total = df_total.drop(columns=['SOLICITACAO','TIPO','ORGAO','OBSERVACAO','REGIONAL_CIDADAO','BAIRRO_CIDADAO','TIPO_CIDADAO','RESPOSTA_FINAL'])

#verificando os dados



df_total = df_total.dropna()

df_total.info()

# Existe um problema na coluna 'DATA_NASC' do solicitante. Existem registros com somente espaos, vamos subistituir esses espaos por uma lista randomica normal 

df_total.loc[df_total['DATA_NASC'] == '                              ', 'DATA_NASC'] = '01/01/1923'

df_total['DATA_NASC'] = pd.to_datetime(df_total['DATA_NASC'])

df_total.to_csv('df_total.csv',index = False)


```
