```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np 

```

```python
df_total = pd.read_csv('df_total.csv')
df_total.head()

# Vamos retirar da amostra registros que tiveram pessoas com menos de 18 anos e pessoas com mais de 100 anos

df_total.drop(df_total[df_total['DATA_NASC'] > '2005-01-01'].index, inplace=True)

df_total.drop(df_total[df_total['DATA_NASC'] < '1922-01-01'].index, inplace=True)

# Proximo passo e criar uma mascara para que possamos analisar a data de nascimento dos requerentes sem a data (1923-01-01) que fizemos para substituir os dados ausentes

mask1 = df_total.loc[df_total['DATA_NASC'] == '1923-01-01']

# df_teste representa o dataframe sem a mascara recem criada  

df_teste = df_total[~df_total.isin(mask1)]

df_teste['DATA_NASC'] = pd.to_datetime(df_teste['DATA_NASC'])



from datetime import datetime

# Encontrando o requerente mais antigo e o mais jovem 

start = (df_teste['DATA_NASC'].sort_values().iloc[0])
end = df_total['DATA_NASC'].sort_values(ascending= False).iloc[0]

d0 = datetime.strptime(str(start), '%Y-%m-%d %H:%M:%S')
d1 = datetime.strptime(str(end), '%Y-%m-%d')

# Calcular a diferenÃ§a de idade (em dias) do cliente mais antigo e do mais jovem

diference = (d1 - d0)
print(diference.days)

#Vamos substituir os valores de 1923-01-01 por valores randomicos de data

import random
import datetime

datas_randomicas = []
for dates in df_total.loc[df_total['DATA_NASC'] == '1923-01-01','DATA_NASC']:
    random_days = random.randint(0,diference.days)
    random_dates = d0 + datetime.timedelta(days = random_days)
    datas_randomicas.append(random_dates)
    


df_total.loc[df_total['DATA_NASC'] == '1923-01-01','DATA_NASC'] = datas_randomicas

# Verificando o que fizemos 

df_total['DATA_NASC'].value_counts()



    

    

        
    




```

```python
df_total['DATA'] = pd.to_datetime(df_total['DATA'])


#Avaliar a coluna sexo

df_total.drop(df_total[df_total['SEXO'] == 'I'].index, inplace=True)
df_total['SEXO'].value_counts()



#Avaliando Orgao Resp

df_total['ORGAO_RESP'].unique()


df_total.sample(n=10)

#Vamos organizar os dfs por tabelas de fato e dimensao para exportar pro PowerBI




df_total['ID'] = np.arange(len(df_total)).astype('str')
df_total['ID'] = df_total['ID'].add('total')

df_assunto = df_total[['ASSUNTO','SUBDIVISAO','DESCRICAO']]
df_cidadao = df_total[['DATA_NASC','SEXO']]
df_local = df_total[['BAIRRO_ASS','REGIONAL_ASS','LOGRADOURO_ASS']]
df_tempo = df_total[['DATA','HORARIO']]



```

```python
df_cidadao = df_cidadao.groupby(['DATA_NASC','SEXO'])
df_cidadao = df_cidadao.all().reset_index()
df_total.head()


```

```python

df_assunto = df_assunto.groupby(['ASSUNTO','SUBDIVISAO','DESCRICAO'])
df_assunto = df_assunto.all().reset_index()
df_assunto.shape


```

```python
df_local = df_local.groupby(['REGIONAL_ASS','BAIRRO_ASS','LOGRADOURO_ASS'])
df_local = df_local.all().reset_index()
df_local.shape

```

```python
df_tempo = df_tempo.groupby(['HORARIO','DATA'])
df_tempo = df_tempo.all().reset_index()
df_tempo.shape

```

```python
df_tempo['ID_tempo'] = np.arange(len(df_tempo)).astype('str')
df_tempo['ID_tempo'] = df_tempo['ID_tempo'].add('T')

df_local['ID_local'] = np.arange(len(df_local)).astype('str')
df_local['ID_local'] = df_local['ID_local'].add('L')

df_assunto['ID_assunto'] = np.arange(len(df_assunto)).astype('str')
df_assunto['ID_assunto'] = df_assunto['ID_assunto'].add('A')

df_cidadao['ID_cidadao'] = np.arange(len(df_cidadao)).astype('str')
df_cidadao['ID_cidadao'] = df_cidadao['ID_cidadao'].add('C')


df_cidadao.head()
```

```python


df_total['DATA_NASC'] = pd.to_datetime(df_total['DATA_NASC'])
df_unido = df_total.merge(df_tempo.set_index(['DATA','HORARIO']),on = ['DATA','HORARIO'], how='left').merge(df_local.set_index(['REGIONAL_ASS','BAIRRO_ASS','LOGRADOURO_ASS']), on = ['REGIONAL_ASS','BAIRRO_ASS','LOGRADOURO_ASS'], how = 'left'
                                                                                                            ).merge(df_assunto.set_index(['ASSUNTO','SUBDIVISAO','DESCRICAO']), on = ['ASSUNTO','SUBDIVISAO','DESCRICAO'], how = 'left'
                                                                                                                    ).merge(df_cidadao.set_index(['DATA_NASC','SEXO']), on = ['DATA_NASC','SEXO'] , how = 'left')



df_unido.head()
```

```python
df_final = df_unido.drop(columns=(['DATA','HORARIO','ASSUNTO','SUBDIVISAO','DESCRICAO','LOGRADOURO_ASS','BAIRRO_ASS','REGIONAL_ASS','MEIO_RESPOSTA','SEXO','DATA_NASC']))
df_final.head()
```
